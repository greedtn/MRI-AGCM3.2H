```python
import func
from csv import reader
import matplotlib.pyplot as plt

with open('../pot_csv/' + 'HFA_KF_c0' + '_POT.csv', 'r') as csv_file:
    csv_reader = reader(csv_file)
    POT_ALL = list(csv_reader)
POT = POT_ALL[2932]
s = []
for i in range(len(POT)):
    s.append(float(POT[i]))
s = sorted(s, reverse=True)
POT = s[:50]
r1, l1 = func.lwm_gpd(data=POT, error=[0.005], thr=POT[-1], period=25, RP=100, n=219143, n0=len(POT), con=0.95, img_name="../png/param/HPA_YS___.png")
with open('../pot_csv/' + 'HPA_KF' + '_POT.csv', 'r') as csv_file:
    csv_reader = reader(csv_file)
    POT_ALL = list(csv_reader)
POT = POT_ALL[2932]
s = []
for i in range(len(POT)):
    s.append(float(POT[i]))
s = sorted(s, reverse=True)
POT = s[:50]
r2, l2 = func.lwm_gpd(data=POT, error=[0.005], thr=POT[-1], period=25, RP=100, n=219143, n0=len(POT), con=0.95, img_name="../png/param/HPA_YS___.png")

# figureを生成する
fig = plt.figure(figsize=(10, 10))
# axをfigureに設定する
ax = fig.add_subplot(1, 1, 1)
# x1, y1を青色でプロット
ax.scatter(r1, l1, c='#aaaaFF')
# x2, y2を青色でプロット
ax.scatter(r2, l2, c='#FFaaaa')
ax.legend(["HFA_YS_c0", "HPA_YS"])
plt.show()
```
